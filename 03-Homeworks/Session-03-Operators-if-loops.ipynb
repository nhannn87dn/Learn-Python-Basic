# Bài tập Session 03 - Luồng điều khiển và Vòng lặp


## 3.1 Input - Đầu Vào



LƯU Ý: CHỈ ĐƯỢC SỬ DỤNG NHỮNG KIẾN THỨC ĐÃ HỌC TỪ SESSION 1 - 3


**Câu 1:** Viết chương trình yêu cầu người dùng nhập tên và tuổi của họ, sau đó in ra "Xin chào **[tên]**, bạn **[tuổi]** tuổi."

**Câu 2:** Viết chương trình yêu cầu người dùng nhập một số nguyên và kiểm tra xem số đó có là số chẵn hay lẻ, sau đó in ra kết quả.

**Câu 3:** Viết chương trình yêu cầu người dùng nhập một số thực và tính diện tích của hình tròn có bán kính là số đó.

**Câu 4:** Viết chương trình yêu cầu người dùng nhập 2 số và in tổng của 2 số đó

**Câu 5:** Viết chương trình yêu cầu người dùng nhập một biểu thức toán học và in ra kết quả

**Câu 6:** Viết chương trình yêu cầu người dùng nhập điểm của một sinh viên và kiểm tra xem sinh viên đó có đạt điểm trung bình (từ 5.0 trở lên) và không có điểm dưới 3.0 không.

**Câu 7:** Viết chương trình yêu cầu người dùng nhập một số nguyên và kiểm tra xem số đó có chia hết cho cả 3 và 5 không.

**Câu 8:** Viết chương trình yêu cầu người dùng nhập một số nguyên và kiểm tra xem số đó có nằm ngoài khoảng từ 1 đến 100 không và đồng thời không là số chia hết cho 5.

**Câu 9:** Viết chương trình yêu cầu người dùng nhập một số và kiểm tra xem số đó có là số âm, dương hay bằng không.

**Câu 10:** Viết chương trình yêu cầu người dùng nhập điểm của một bài kiểm tra và in ra "Đậu" nếu điểm lớn hơn hoặc bằng 5 và in ra "Trượt" nếu điểm nhỏ hơn 5.

**Câu 11:** Viết chương trình yêu cầu người dùng nhập ba số và in ra số lớn nhất trong ba số đó.

**Câu 12:** Viết chương trình yêu cầu người dùng nhập một số nguyên và kiểm tra xem số đó có là số chia hết cho 3 và 5 đồng thời không là số chia hết cho 2 hay không

**Câu 13:** Viết chương trình yêu cầu người dùng nhập năm sinh của họ và in ra "Bạn là người trưởng thành" nếu tuổi của họ lớn hơn hoặc bằng 18, và in ra "Bạn là người vị thành niên" nếu tuổi của họ nhỏ hơn 18.

**Câu 14:** Kiểm tra tuổi để vào một trang web

- Viết một chương trình yêu cầu người dùng nhập tuổi của họ.

- Nếu người dùng dưới 18 tuổi, in ra thông báo "Bạn chưa đủ tuổi để truy cập trang web này."

- Nếu người dùng từ 18 tuổi trở lên, in ra thông báo "Chào mừng bạn đến với trang web!"

**Câu 15:** Viết một chương trình yêu cầu người dùng nhập tình trạng thời tiết hiện tại (nắng, mưa, hoặc có mây)"
- In ra thông báo phù hợp dựa trên tình trạng thời tiết:
 - "Hôm nay trời nắng, đừng quên đeo kính râm!"
 - "Hôm nay trời mưa, hãy mang theo ô!"
 - "Hôm nay trời có mây, thời tiết thật dễ chịu!"
- Nếu người dùng nhập bất kỳ giá trị nào khác, in ra thông báo "Tình trạng thời tiết không xác định."

**Câu 16:** Viết một chương trình yêu cầu người dùng nhập mật khẩu.
- Nếu mật khẩu đúng (ví dụ: "python123"), in ra thông báo "Mật khẩu đúng, truy cập thành công."
- Nếu mật khẩu sai, in ra thông báo "Mật khẩu sai, vui lòng thử lại."

**Câu 17:** Viết một chương trình yêu cầu người dùng nhập loại phương tiện họ đang sử dụng (xe đạp, xe máy, ô tô).
- In ra thông báo phù hợp dựa trên loại phương tiện:
  - "Bạn đang đi xe đạp, hãy đội mũ bảo hiểm."
  - "Bạn đang đi xe máy, hãy đeo mũ bảo hiểm và áo giáp."
  - "Bạn đang lái ô tô, hãy thắt dây an toàn."
- Nếu người dùng nhập bất kỳ giá trị nào khác, in ra thông báo "Loại phương tiện không xác định."

## 3.4 Vòng lặp

**Câu 1:** Viết chương trình Python in ra các số từ 1 đến 10

**Câu 2:** Viết chương trình Python tính tổng các số từ 1 đến n, với n là một số nguyên được nhập từ người dùng

**Câu 3:** Viết chương trình Python in ra bảng cửu chương từ 2 đến 9

**Câu 4:** Viết chương trình Python in ra các ký tự trong một chuỗi được nhập từ người dùng, mỗi ký tự trên một dòng

**Câu 5:** Viết chương trình Python tính tổng các phần tử trong một danh sách số nguyên được nhập từ người dùng.

**Câu 6:** Viết chương trình Python tính tổng các số chẵn từ 1 đến n, với n là một số nguyên được nhập từ người dùng

**Câu 7:** Viết chương trình Python in ra các số nguyên tố từ 2 đến n, với n là một số nguyên được nhập từ người dùng.

**Câu 8:** Viết chương trình Python kiểm tra giá trị nhập vào của người dùng. Tên thủ đô của Nhật Bản là gì ? Nếu giá trị nhập vào không phải là **'Tokyo'** thì hiển thị thông báo.
- Câu trả lời của bạn chưa chính xác. Vui lòng nhập lại.
- Bạn tạo input để cho người dùng nhập lại câu trả lời.
