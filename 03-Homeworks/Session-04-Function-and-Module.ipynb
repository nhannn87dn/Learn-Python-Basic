# Bài tập Session 04 - Hàm và Module

## Phần 1. Module

**Câu 1:** Viết chương trình Python để tính diện tích của hình tròn, sử dụng hằng số PI và các hàm trong module `math`.

**Câu 2:** Viết chương trình Python để in ra ngày và giờ hiện tại, sử dụng module `datetime`

*❗ Lưu ý quan trọng:*

> *- Trong Colab, những module bạn đã import từ phần trên rồi, thì kể từ đó bên dưới không cần import lại nữa. Colab tự hiểu để sử dụng lại.*



**Câu 3:** Viết chương trình Python để tạo một danh sách gồm 5 số ngẫu nhiên trong khoảng từ 1 đến 100, sử dụng module `random`.

**Câu 4:** Viết chương trình Python để liệt kê tất cả các tập tin trong thư mục hiện tại, sử dụng module `os`.

**Câu 5:** Viết chương trình Python để tải nội dung của một trang web bất kỳ, sử dụng module `requests`.

## Phần 2: Function - Hàm

**Câu 1:** Viết một hàm có tên **calculate_sum** dưới dạng `module`,  nhận vào hai tham số và trả về tổng của hai số đó.

**Câu 2:** Viết một hàm có tên **reverse_string** nhận vào một chuỗi và trả về chuỗi đảo ngược của chuỗi đó.


**Câu 3:** Viết một hàm có tên **find_max** nhận vào một danh sách các số và trả về số lớn nhất trong danh sách đó

**Câu 4:** Viết một hàm có tên **calculate_power** dưới dạng `module` nhận vào 3 tham số: `x`, `y`, `type`. Trong đó:
- `x`: là một số
- `y`: là một số
- `phep_tinh`: là một chuỗi thuộc một trong các giá trị: `cong`, `tru`, `nhan`, `chia`

Hàm này sẽ trả về kết quả tương ứng với phép tính đã được chỉ định.

**Câu 5:** Viết một hàm có tên get_discount nhận vào một tham số duy nhất là **membership_level** (level thành viên) và trả về phần trăm chiết khấu tương ứng với level thành viên đó. Các level thành viên và phần trăm chiết khấu tương ứng như sau:

- Bronze: 5%
- Silver: 10%
- Gold: 15%
- Platinum: 20%
- Nếu **membership_level** không khớp với bất kỳ level nào trên, hàm sẽ trả về 0%

**Câu 6:** Viết một hàm có tên **get_bus_fare** nhận vào một tham số duy nhất là ticket_type (loại vé) và trả về giá vé tương ứng. Các loại vé và giá vé tương ứng như sau:

- **normal**: 7000 VND
- **mstudent**: 50000 VND
- **madult**: 100000 VND
- Nếu ticket_type không khớp với bất kỳ loại vé nào trên, hàm sẽ trả về thông báo "Loại vé không hợp lệ".

**Câu 7:** Viết một hàm có tên **get_drink_price** nhận vào một tham số duy nhất là ***drink_type*** (loại đồ uống) và trả về giá tiền tương ứng. Các loại đồ uống và giá tiền tương ứng như sau:

- water: 5000 VND
- tea: 30000 VND
- coffe: 20000 VND
- Nếu drink_type không khớp với bất kỳ loại đồ uống nào trên, hàm sẽ trả về thông báo "Loại đồ uống không hợp lệ".

**Câu 8:** Viết một hàm có tên **get_grade** nhận vào một tham số duy nhất là score (điểm số) và trả về xếp loại học lực tương ứng. Các mức xếp loại như sau:

- 90 <= score <= 100: "Xuất sắc"
- 80 <= score < 90: "Giỏi"
- 70 <= score < 80: "Khá"
- 50 <= score < 70: "Trung bình"
- score < 50: "Yếu"

**Câu 9:** Viết một hàm có tên get_time_of_day nhận vào một tham số duy nhất là hour (giờ trong ngày, từ 0 đến 23) và trả về khoảng thời gian tương ứng. Các khoảng thời gian như sau:

- 0 <= hour < 12: "Buổi sáng"
- 12 <= hour < 18: "Buổi chiều"
- 18 <= hour < 21: "Buổi tối"
- 21 <= hour < 24: "Buổi đêm"

**Câu 10:** Viết một hàm có tên **calculate_shipping_fee** nhận vào hai tham số **distance** (khoảng cách giao hàng tính theo km) và **weight** (trọng lượng hàng hóa tính theo kg), trả về phí giao hàng tương ứng dựa trên các quy tắc sau:

- Phí cơ bản dựa trên khoảng cách:
  - 0 km <= distance <= 5 km: 10000 VND
  - 5 km < distance <= 20 km: 20000 VND
  - distance > 20 km: 50000 VND
- Phụ phí dựa trên trọng lượng:
  - weight <= 2 kg: 0 VND (không phụ phí)
  - 2 kg < weight <= 5 kg: 5000 VND
  - weight > 5 kg: 10000 VND
- Phí giao hàng tổng cộng là tổng của phí cơ bản và phụ phí.
