Cách tạo folder:

Bước 1: Click vào biểu tượng **Folder** như hình dưới (Chỗ khoanh tròn 1)

Bước 2: Click phải lên vùng số 2 (chỗ khoanh tròn 2) --> Chọn **New Folder**. Đặt tên là **homeworks**.

Bước 3: Click phải lên folder **homeworks** --> chọn **Upload**. Sau đó  Upload tập tin demo-1.txt, demo-2.txt vào folder homeworks
