# Bài tập Session 05 - Class OOP - Lập trình hướng đối tượng

**Câu 1:** Tạo một lớp có tên `Dog` với các thuộc tính và phương thức sau:

- Thuộc tính:
  - `name` (tên của chó)
  - `age` (tuổi của chó)
- Phương thức:
  - `bark()`: In ra màn hình chuỗi "Woof!".
  - `get_info()`: Trả về một chuỗi chứa tên và tuổi của chó.

- Sau đó khởi tạo 2 đối tượng chó `Husky` và `Bull`, Sử dụng phương thức đã định nghĩa để In ra thông tin mỗi đối tượng tương ứng.

**Câu 2:** Định nghĩa một lớp `Student`

Tạo một lớp có tên `Student` với các thuộc tính và phương thức sau:

- Thuộc tính:
  - `name` (tên của học sinh)
  - `grade` (lớp của học sinh)
  - `scores` (điểm số của học sinh, là một danh sách các số)
- Phương thức:
  - `add_score(score)`: Thêm một điểm vào danh sách scores.
  - `get_average()`: Tính và trả về điểm trung bình của học sinh.

- Sau đó khởi tạo 2 đối tượng học sinh `hocSinhA` và `hocSinhB`, Sử dụng phương thức đã định nghĩa để In ra thông tin mỗi đối tượng tương ứng.

**Câu 3:** Định nghĩa một lớp `Book`

Tạo một lớp có tên `Book` với các thuộc tính và phương thức sau:

- Thuộc tính:
  - `title` (tên sách)
  - `author` (tác giả)
  - `num_pages` (số trang)
  - `price` (Giá bìa)
- Phương thức:
  -  `get_description()`: Trả về một chuỗi chứa thông tin về sách.
  - `get_price()`: Trả về giá bìa của sách

- Sau đó khởi tạo 2 đối tượng sách `TruyenTranh` và `TinHoc`, Sử dụng phương thức đã định nghĩa để In ra thông tin mỗi đối tượng tương ứng.

**Câu 4:** Định nghĩa một lớp `Car`

Tạo một lớp có tên `Car` với các thuộc tính và phương thức sau:

- Thuộc tính:
  - `make` (hãng sản xuất)
  - `model` (mẫu xe)
  - `year` (năm sản xuất)
  - `wheels` (Số bánh)
  - `seats` (Số chỗ ngồi)
- Phương thức:
  - `start()`: In ra màn hình chuỗi "The car is starting.".
  - `get_info()`: Trả về một chuỗi chứa thông tin về xe.

- Sau đó khởi tạo 2 đối tượng xe `Taxi` và `Bus`, Sử dụng phương thức đã định nghĩa để In ra thông tin mỗi đối tượng tương ứng.

**Câu 5:** Tạo một lớp có tên `GameCharacter` với các thuộc tính và phương thức sau:

1. **Thuộc tính:**
   - `name` (tên nhân vật)
   - `level` (cấp độ của nhân vật)
   - `health` (sức khỏe của nhân vật)

2. **Phương thức:**
   - `level_up()`: Tăng cấp độ của nhân vật lên 1.
   - `attack()`: Trả về một chuỗi mô tả hành động tấn công dựa trên cấp độ của nhân vật.
   - `defend()`: Trả về một chuỗi mô tả hành động phòng thủ dựa trên cấp độ của nhân vật.

**Yêu cầu chi tiết:**

- Khi nhân vật ở cấp độ 1, hành động tấn công chỉ là `"Basic Attack"`, và hành động phòng thủ là `"Basic Defense"`.
- Khi nhân vật ở cấp độ 2, hành động tấn công là `"Power Strike"`, và hành động phòng thủ là `"Shield Block"`.
- Khi nhân vật ở cấp độ 3 trở lên, hành động tấn công là `"Ultimate Slash"`, và hành động phòng thủ là `"Magic Barrier"`.

3. **Khởi tạo đối tượng:**

- Sau đó khởi tạo 2 đối tượng nhân vật Game `NhanVatA` và `NhanVatB`, Sử dụng phương thức đã định nghĩa để In ra thông tin mỗi đối tượng tương ứng.
